```python
from sklearn import linear_model
from sklearn.datasets import load_diabetes, load_boston
from sklearn.linear_model import RidgeCV
import numpy as np
from matplotlib import pyplot as plt
```

```python
X, y = load_diabetes(return_X_y=True)
alphas = np.arange(0.01, 2, 0.01)
clf = RidgeCV(alphas=alphas, fit_intercept=False, store_cv_values=True).fit(X, y)
cvs = np.mean(clf.cv_values_, axis=0)
plt.plot(alphas, cvs)
plt.ylabel("estimated prediction error")
plt.xlabel("位")
plt.yticks([])
plt.title("Optimal 位 from cross validation")
plt.show()
```

```python
X, y = load_diabetes(return_X_y=True)
XX = X - np.mean(X, axis=0) # centered
XXX = XX/np.std(XX, axis=0) # scaled
print(np.std(XXX, axis=0))
alphas = np.arange(0.1, 1000, 1)
clf = RidgeCV(alphas=alphas, fit_intercept=False, store_cv_values=True).fit(XXX, y)
cvs = np.mean(clf.cv_values_, axis=0)
plt.plot(alphas, cvs)
plt.ylabel("estimated prediction error")
plt.xlabel("位")
plt.yticks([])
plt.title("Optimal 位 from cross validation")
plt.show()
```
